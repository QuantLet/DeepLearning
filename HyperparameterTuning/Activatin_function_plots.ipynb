```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Generate input values
x = np.linspace(-10, 10, 400)

# Compute the outputs of each activation function
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)

# Plot Sigmoid
plt.figure(figsize=(10,6))
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.plot(x, y_sigmoid, color='blue')
plt.xticks(np.arange(-10, 11, 2), fontsize=18)  # Integer ticks for x-axis
plt.yticks(np.arange(0, 2, 0.5)[:-1], fontsize=18)     # Integer ticks for y-axis (0 and 1 for Sigmoid range)
plt.savefig('sigmoid.png', transparent=True)
plt.show()

# Plot Tanh
plt.figure(figsize=(10,6))
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.plot(x, y_tanh, color='green')
plt.xticks(np.arange(-10, 11, 2), fontsize=18)  # Integer ticks for x-axis
plt.yticks(np.arange(-1, 2, 0.5)[:-1],fontsize=18)     # Integer ticks for y-axis (-1 and 1 for tanh range)
plt.savefig('tanh.png', transparent=True)
plt.show()

# Plot ReLU
plt.figure(figsize=(10,6))
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.plot(x, y_relu, color='red')
plt.xticks(np.arange(-10, 11, 2), fontsize=18)  # Integer ticks for x-axis
plt.yticks(np.arange(0, 11, 2), fontsize=18)     # Integer ticks for y-axis (0 and 10 for ReLU range)
plt.savefig('ReLU.png', transparent=True)
plt.show()

plt.tight_layout()
plt.show()
```
